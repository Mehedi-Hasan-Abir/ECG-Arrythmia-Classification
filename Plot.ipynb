```python
import numpy as np
import matplotlib.pyplot as plt

# data to plot
n_groups = 4
P = (97.8, 91.9, 87.6, 90.1)
P1 = (98.5, 92.1, 89.2, 90.3)

V = (98.8, 98.0, 97.3, 97.7)
V1 = (98.6, 97.9, 96.9, 96.3)
P = tuple(i/100 for i in P)
P1 = tuple(i/100 for i in P1)
V = tuple(i/100 for i in V)
V1 = tuple(i/100 for i in V1)
# create plot
fig, ax = plt.subplots(figsize =(12, 18))
index = np.arange(n_groups)
bar_width = 0.18
opacity = 0.5

rects1 = plt.bar(index, P, bar_width,
alpha=opacity,
color='purple',
label='Proposed Model')

rects11 = plt.bar(index + bar_width , P1, bar_width,
alpha=.2,
color='red',
label='Proposed Model Augmented')

rects2 = plt.bar(index + bar_width + bar_width, V, bar_width,
alpha=opacity,
color='green',
label='VGG')

rects22 = plt.bar(index + bar_width + bar_width + bar_width, V1, bar_width,
alpha=.2,
color='blue',
label='VGG Augmented')

plt.xlabel('Metrics')
plt.ylabel('Scores')
plt.title('Metrics Evaluation of Prosposed Model and VGG')
plt.xticks(index + bar_width, ('Accuracy', 'Precision', 'Sensivity', 'F1 Score'))
plt.legend()

plt.tight_layout(h_pad=1,pad=1.08,w_pad=5)
plt.show()
```
